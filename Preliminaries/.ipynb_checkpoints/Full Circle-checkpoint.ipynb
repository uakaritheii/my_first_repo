```python
import numpy as np
import matplotlib.pyplot as plt


def create_serp(n):
    x = np.zeros(n)
    y = np.zeros(n)
    for i in range(1,n):
        k = np.random.randint(1,4)
        x[i] = 0.5*x[i-1] + k -1
        y[i] = 0.5*y[i-1]

        if k == 2:
            y[i] += 2

    return x, y

n = 20000
x, y = create_serp(n)
plt.plot(x,y,'.')
        
```
