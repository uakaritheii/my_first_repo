```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2,100)
plt.plot(x,np.sqrt(x))
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([x[0], x[-1]]) # soft coded limit
plt.ylim([0,8])
plt.gca().set_aspect('auto') #gca means get current axes. Does the same thing as above
plt.gca().set_aspect(1/plt.gca().get_data_ratio()) # a good way to force your graph to be square
plt.title("A really awesome plot")  
plt.legend(['y=sqrt(x)', 'y=x**2', 'y=x**3']) #instead of "labels = " parameter inside plot, we can use legends as a catch-all}
plt.grid()
plt.show()
```

```python
#The syntax is different when working with subplots

fig, ax = plt.subplots(1,3,figsize = (10,8))
y = [np.sqrt(x), x**2, x**3]
titles = ['square root', 'square', 'cube']
legends = ['y=sqrt(x)', 'y=x**2', 'y=x**3']
colors = ['k', 'b', 'r']
i = 0

for p in ax:
    p.plot(x, y[i], label = legends[i], color = colors[i])
    p.set_xlim([x[0], x[-1]])
    p.set_ylim([0, y[i][-1]])
    p.set_aspect(1/p.get_data_ratio())
    p.set_title(titles[i])
    p.legend()
    p.grid()
    i+=1
plt.show()
```
