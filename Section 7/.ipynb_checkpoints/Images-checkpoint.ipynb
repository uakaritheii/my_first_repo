```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
M = np.random.randint(10,size=(3,5)) #populates a 3x5 matrix with integers from 0 to 9
print(M)
plt.imshow(M)

for i in range(3):
    for j in range(5):
        plt.text(j, i,str(M[i,j]), horizontalalignment = 'center', fontsize = 30, verticalalignment = 'center') #first parameter is x and second parameter is y
plt.colorbar()
plt.show()
```

```python
H = np.zeros([10,10]) #9 rows 10 columns
len(H[0,:]) # number of columns based on the 0th row
len(H[:,0]) # number of rows based on the 0th column
plt.imshow(H)

#H[i,j] refers to the element at the ith row and jth column
```

```python
for i in range(len(H[:,0])):
    for j in range(len(H[0,:])):
        H[i,j] = 1/(i+j+1)
np.round(H,2)
sns.heatmap(H,vmin=0,vmax=.3)
plt.show()

```
