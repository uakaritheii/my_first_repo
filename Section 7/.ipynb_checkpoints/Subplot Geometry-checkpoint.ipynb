```python
import matplotlib.pyplot as plt
import numpy as np

#subplot geometry with 1 row 2 columns
y = np.random.randn(10)
z = np.random.randn(10)
plt.subplot(1,2,1) #creates a subplot with 1 row, 2 columns and initializes the first
plt.plot(y)

plt.subplot(1,2,2)
plt.plot(z)
```

```python
# How about something more complicated

for i in range(9):
    plt.subplot(3,3,i+1)
    plt.plot(np.random.randn(10))
```

```python
#Control over subplots

pts = np.random.randn(2,2)
print(pts)
print(pts[0,0]) #another indexing method
plt.plot([[3,4],[5,6]]) #creates two plots with the yvals [3,5] and [4,6]. Creates its own x points [0,1] in this case
```

```python
#Further control over subplots

fig, ax = plt.subplots(2,2, figsize=(10,10)) #2x2 subplot matrix. subplotS not subplot
ax[0][0].plot(np.random.randn(4,4))
ax[0,1].plot(np.random.randn(4,4))
ax[1,0].plot(np.random.randn(4,4))
ax[1,1].plot(np.random.randn(4,4))
```

```python
#Exercise: Create a 3x3 subplot geometry and populate the subplots using a for loop

fig, ax = plt.subplots(2,3, figsize = (10,5))
big, bx = plt.subplots(3,2, figsize = (10,8))

#Method 1
for i in range(len(ax)):
    for j in range(len(ax[i])):
        ax[i,j].plot(np.random.randn(2,2))

#Method 2
for i in range(len(bx.flatten())):
    bx.flatten()[i].plot(np.random.randn(2,2))
    
```
