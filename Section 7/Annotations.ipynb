```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5,6) #np arange
y = x**2

fig, ax = plt.subplots()

ax.plot(x,y,'ko-', markerfacecolor = 'w', markersize = 15)
ax.annotate('Hello',(x[3],y[3]), fontsize=15, arrowprops= dict(),xycoords='data',xytext = (0,10), 
            horizontalalignment = 'center')
plt.show()
```

```python
#Exercise: Reporduction

y = x**2 + np.cos(x)*10

#first find the minimum point
minpnt = np.argmin(y) # returns the index of the point where the minimum point lies
txt = "minpoint: (%g, %.2f)"%(x[minpnt], y[minpnt])
print(txt)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y,'ko-', markerfacecolor = 'purple', markersize = 15)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.annotate(txt, (x[minpnt], y[minpnt]), arrowprops=dict(color = 'purple', arrowstyle = 'wedge'),xycoords = 'data', 
            fontsize = 15, xytext = (x[int(len(x)/2)], np.max(y)*0.8), horizontalalignment = 'center')
plt.show()
```
