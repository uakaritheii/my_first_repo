```python
import matplotlib.pyplot as plt

plt.plot(3,4,'o') 
plt.plot(2,4,'s') # The plots are held on by default. A new plot is not created
plt.show() # this essentially locks the plot

plt.plot(20,40,'o')
plt.plot(30,30, 's')
plt.show()
```

```python
plt.plot(3,2,'ro', label = 'red circle')
plt.plot(3,4,'bp', label = 'blue pentagon')
plt.legend() # the legend method has to be explicity called
plt.show() 
```

```python
#Plotting lines
ys = [0,1] # specifies the starting and ending y coord
xs = [0,2] # specifies the starting and ending x coord

plt.plot(xs,ys,'r')
plt.plot([0,2],[1,2], 'g')
plt.legend(['red', 'green'])
plt.show()
```

```python
#numpy can be integrated with matplotlib

import numpy as np

x = np.linspace(0, 3*np.pi, 200)
y = np.sin(x)

plt.plot(x,y,'k.') 
```
