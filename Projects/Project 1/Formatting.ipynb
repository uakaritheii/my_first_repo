```python
# Importing the text file
# NOTE: When working with files always use "with...as" as it prevents having to use txt.close

with open('captions_text.vtt', 'r') as k:
    txt = k.readlines()[4:]

txt
```

```python
#First solution
#This solution is less than ideal as it doesn't eliminate patterns and is much more brute-force
import string

edp = [c for c in string.punctuation]
edp.remove(".")
edp.remove("'")
edp.remove(",")
edp.remove(";")
edp.remove("-")

def ref(text):
    nlst = []

    for i in range(len(text)):
        toggle = True
        if text[i].strip().isnumeric() or text[i].strip() == '':
            toggle = False
            continue
        for c in edp:
            if c in text[i]:
                toggle = False
                continue
        if toggle:
            nlst.append(text[i].strip())

    return ' '.join(nlst)


ref(txt)
```

```python
#A brief note on re

import re

sent = "Hi, my name is Mike and I like to eat Mike's chocolate"

regex_name = re.compile('Mike') #tells the re module the string we're searching for
regex_name.findall(sent) #tells us there are two instances of Mike (CASE SENSITIVE)
regex_name.sub('barry', sent)


# What if we want to eliminate patterns and not just specific words

cent = "Hi, my name is Mike and I like to eat Moke's chocolate"

regex_name = re.compile(r'M\wke')
regex_name.findall(cent)
regex_name.sub('', cent)
txt
```

```python
def reformat(text):
    rennn = re.compile(r'\d{1,3}\n|\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d|\n') #d{1,3} tells us to remove 1,2, and 3 digit combos

    text = rennn.sub('', text)

    return text

newstr  = ' '.join([line.strip() for line in [reformat(line) for line in txt] if line])

newstr
```

```python
with open('edit.txt', 'w') as f:
    f.write(newstr)
        
```
